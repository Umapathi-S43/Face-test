## ü§ñ Step 2: Download AI Models

```python
import os
from pathlib import Path

# Create models directory
models_dir = Path("app/models")
models_dir.mkdir(exist_ok=True, parents=True)

print("üì• Downloading AI models (650MB total)...")

# Download InSwapper model
if not (models_dir / "inswapper_128.onnx").exists():
    print("  ‚¨áÔ∏è InSwapper model (300MB)...")
    !wget -q --show-progress -O app/models/inswapper_128.onnx \
      "https://huggingface.co/hacksider/deep-live-cam/resolve/main/inswapper_128.onnx"
    print("  ‚úÖ InSwapper downloaded")

# Download GFPGAN model
if not (models_dir / "GFPGANv1.4.pth").exists():
    print("  ‚¨áÔ∏è GFPGAN model (350MB)...")
    !wget -q --show-progress -O app/models/GFPGANv1.4.pth \
      "https://github.com/TencentARC/GFPGAN/releases/download/v1.3.4/GFPGANv1.4.pth"
    print("  ‚úÖ GFPGAN downloaded")

print("\n‚úÖ All models ready!")
```
